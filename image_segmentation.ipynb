```python
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image = Image.open('anirudh2.jpeg')
plt.imshow(image)
```

```python
image = np.array(image)
```

```python
original_size = image.shape
```

```python
image = image.reshape((-1,3))
print(image.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=10)
kmeans.fit(image)
```

```python
centers = np.array(kmeans.cluster_centers_,dtype='uint8')
```

```python
plt.subplot?
```

```python
i = 1

plt.figure(0,figsize=(8,2))


colors = []

for each_col in centers:
    plt.subplot(3,10,i)
    plt.axis("off")
    i+=1
    
    colors.append(each_col)
    
    #Color Swatch
    a = np.zeros((100,100,3),dtype='uint8')
    a[:,:,:] = each_col
    plt.imshow(a)
    
plt.show()
```

```python
size = image.shape[0]
#size = new_image.shape[0]*new_image.shape[1]
new_image = np.zeros((size,3),dtype='uint8')
new_image.shape
```

```python
for ix in range(new_image.shape[0]):
    new_image[ix] = colors[kmeans.labels_[ix]]
    
new_img = new_image.reshape((original_size))
plt.imshow(new_img)
plt.show()
    
    
```

```python
plt.imsave("anirudh_modified.jpg",new_img)
```
